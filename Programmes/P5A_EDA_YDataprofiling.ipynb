# PROJET 5A - Analyse Exploratoire Automatisée (Data Profiling)

**Objectif :** Réaliser un audit technique de l'ensemble des jeux de données générés lors de l'étape de préparation.

**Pourquoi utiliser le Profiling Automatisé ?**
`ydata-profiling` scanne systématiquement chaque variable pour détecter :
1.  **Qualité des données :** Taux de valeurs manquantes, zéros, doublons.
2.  **Distributions :** Skewness (asymétrie), Kurtosis, outliers.
3.  **Corrélations :** Matrice complète (Pearson, Spearman) pour identifier la multicolinéarité.
4.  **Interactions :** Croisement des variables deux à deux.

**Périmètre de l'analyse :**
Nous générons un rapport distinct pour chaque sous-ensemble afin de valider nos stratégies de découpage :
* `Select_Championnats` : Le dataset final pour la modélisation (Qualité maximale).
* `Recent_Data` : Pour vérifier la stabilité des distributions sur la période récente.
* `Matches_with_Elo` : Le dataset maître (Audit global).
* `Clean` & `Pred_Full` : Pour comparaisons techniques.

**Sortie :**
Les rapports sont générés au format **HTML** interactif

```python
import sys
import os
import pandas as pd
from ydata_profiling import ProfileReport

sys.path.append(os.getcwd())
try:
    import config as cfg
    print("Configuration chargée")
except ImportError:
    print("Config introuvable")

# liste datasets
datasets_to_profile = {
    "Select_Championnats": cfg.FILE_SELECT,  
    "Recent_Data": cfg.FILE_RECENT,     
    "Matches_with_Elo": cfg.FILE_MATCHES_ELO,        
    "matches_clean": cfg.FILE_CLEAN,                      
    "matches_pred_full": cfg.FILE_PRED_FULL       
}

# BOUCLE 
for name, path in datasets_to_profile.items():
    print(f"\nDémarrage du profiling pour : {name}")
    
    if not path.exists():
        print(f"Fichier introuvable : {path}")
        continue
     
    df = pd.read_csv(path, sep=";", parse_dates=["MatchDate"], dayfirst=True)
    print(f"   Dimensions : {df.shape}")

    # Config
    profile = ProfileReport(
        df, 
        title=f"Rapport EDA - {name}", 
        explorative=True,
        minimal=len(df) > 100000 
    )

    # HTML
    output_file = cfg.DATA_DIR_S / f"Report_{name}.html"
    profile.to_file(output_file)
    print(f"Rapport HTML sauvegardé : {output_file}")
```
