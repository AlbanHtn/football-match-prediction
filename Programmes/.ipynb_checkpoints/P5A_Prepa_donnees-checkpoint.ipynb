## Features Engineering 

```python
print(matches.columns.tolist())
print(elo.columns.tolist())
```

```python
matches = pd.read_csv("C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/Matches.csv", 
                      sep=";", parse_dates=["MatchDate"], dayfirst=True, low_memory=False)
elo = pd.read_csv("C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/EloRatings.csv", 
                  sep=";", parse_dates=["date"], dayfirst=True)

elo = elo.sort_values(["club", "date"]).reset_index(drop=True)

def compute_elo_change(elo_df, months=1):
    ''' 
    Calcul variation du classement Elo pour chaque club sur différentes périodes (1, 3, 6 mois)
    -> mesurer la dynamique de performance
    '''
    col_name = f"EloChange{months}M"
    elo_df[col_name] = elo_df.groupby("club")["elo"].diff(periods=months*2)
    return elo_df

for m in [1, 3, 6]:
    elo = compute_elo_change(elo, months=m)

elo_features = ["club", "date", "elo", "country", "EloChange1M", "EloChange3M", "EloChange6M"]
elo = elo[elo_features]

matches["MatchDate"] = pd.to_datetime(matches["MatchDate"], dayfirst=True, errors="coerce")
elo["date"] = pd.to_datetime(elo["date"], dayfirst=True, errors="coerce")

def merge_asof_team(matches, elo, team_col, prefix):
    """
    obligatoire pour association elo récent d'avant match
    """
    out = []
    for team, df_team in matches.groupby(team_col):
        df_team = df_team.sort_values("MatchDate")
        df_elo = elo[elo["club"] == team].sort_values("date")
        merged = pd.merge_asof(
            df_team,
            df_elo.rename(columns={
                "club": team_col,
                "elo": f"{prefix}EloSnap",
                "EloChange1M": f"{prefix}EloChange1M",
                "EloChange3M": f"{prefix}EloChange3M",
                "EloChange6M": f"{prefix}EloChange6M"
            }),
            by=team_col,
            left_on="MatchDate",
            right_on="date",
            direction="backward",
            allow_exact_matches=True
        )
        out.append(merged)
    return pd.concat(out, ignore_index=True)

# Merge 
matches = merge_asof_team(matches, elo, "HomeTeam", "Home")
matches = merge_asof_team(matches, elo, "AwayTeam", "Away")

matches.drop(columns=["date"], inplace=True, errors="ignore")

# momentum
matches["EloChange1M_Diff"] = matches["HomeEloChange1M"] - matches["AwayEloChange1M"]
matches["EloChange3M_Diff"] = matches["HomeEloChange3M"] - matches["AwayEloChange3M"]
matches["EloChange6M_Diff"] = matches["HomeEloChange6M"] - matches["AwayEloChange6M"]

# elo
matches["EloDiff"] = matches["HomeElo"] - matches["AwayElo"] 
matches["EloTotal"] = matches["HomeElo"] + matches["AwayElo"] 
matches["EloAdvantage"] = matches["EloDiff"] / matches["EloTotal"]

matches["Form3Diff"] = matches["Form3Home"] - matches["Form3Away"]
matches["Form5Diff"] = matches["Form5Home"] - matches["Form5Away"]
matches["FormMomentumHome"] = matches["Form3Home"] - (matches["Form5Home"] - matches["Form3Home"])
matches["FormMomentumAway"] = matches["Form3Away"] - (matches["Form5Away"] - matches["Form3Away"])

# Probabilités (odds)
matches["PHome"] = 1 / matches["OddHome"]
matches["PDraw"] = 1 / matches["OddDraw"]
matches["PAway"] = 1 / matches["OddAway"]

# Normalisation
matches["ProbSum"] = matches["PHome"] + matches["PDraw"] + matches["PAway"]
matches["PHome_norm"] = matches["PHome"] / matches["ProbSum"]
matches["PDraw_norm"] = matches["PDraw"] / matches["ProbSum"]
matches["PAway_norm"] = matches["PAway"] / matches["ProbSum"]

matches["ShotsDifference"] = matches["HomeShots"] - matches["AwayShots"]
matches["CornersDifference"] = matches["HomeCorners"] - matches["AwayCorners"]
matches["CardsHome"] = matches["HomeYellow"] + 2*matches["HomeRed"]
matches["CardsAway"] = matches["AwayYellow"] + 2*matches["AwayRed"]
matches["CardsDiff"] = matches["CardsHome"] - matches["CardsAway"]
matches["GameDominanceIndex"] = (matches["ShotsDifference"] + matches["CornersDifference"]) / 2

print("\nComplétude des Elo après merge_asof :")
print(matches[["HomeEloSnap", "AwayEloSnap"]].notna().mean() * 100)
```

```python
from rapidfuzz import process

divmap = pd.read_csv("C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/division_to_country_TEMPLATE.csv", sep=";")
elo_full = pd.read_csv("C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/EloRatings.csv", sep=";")

matches["Country"] = matches["Division"].map(dict(zip(divmap["Division"], divmap["CountryCode"])))

# On garde seulement les matchs dont le pays est couvert par Elo
matches_elo = matches[matches["Country"].isin(elo_full["country"].unique())].copy()

print(f"Matchs initiaux : {len(matches)}")
print(f"Matchs avec Elo disponible : {len(matches_elo)} ({len(matches_elo)/len(matches):.1%})")

missing_clubs = set(matches[matches["HomeEloSnap"].isna()]["HomeTeam"].unique()) | \
                set(matches[matches["AwayEloSnap"].isna()]["AwayTeam"].unique())
```

```python
# 5️⃣ Fuzzing amélioré avec filtrage par pays
elo_clubs = elo["club"].unique()
suggestions = []

for club in sorted(missing_clubs):
    # Pays probables selon la division du club
    possible_countries = matches.loc[matches["HomeTeam"]==club, "Country"].dropna().unique().tolist()
    if not possible_countries:
        possible_countries = matches.loc[matches["AwayTeam"]==club, "Country"].dropna().unique().tolist()
    
    # Si on a trouvé des pays, on filtre
    if possible_countries:
        possible_elo = elo[elo["country"].isin(possible_countries)]["club"].unique()
    else:
        possible_elo = elo_clubs

    # Recherche des meilleures correspondances dans le bon pays
    best_matches = process.extract(club, possible_elo, limit=3)
    for match, score, _ in best_matches:
        suggestions.append({
            "club_missing": club,
            "suggested_match": match,
            "similarity": score,
            "countries_considered": ", ".join(possible_countries) if possible_countries else "ALL"
        })
```

```python
# 6️⃣ Sauvegarde et filtrage des mappings
suggestions_df = pd.DataFrame(suggestions)

# On retire les suggestions dont le pays n'a pas pu être déterminé (countries_considered == "ALL" ou vide)
suggestions_df = suggestions_df[
    (suggestions_df["countries_considered"].notna()) &
    (suggestions_df["countries_considered"] != "ALL") &
    (suggestions_df["countries_considered"].str.strip() != "")
]

# Tri et sélection du meilleur match pour chaque club manquant
best_suggestions = (
    suggestions_df.sort_values(["club_missing", "similarity"], ascending=[True, False])
    .groupby("club_missing")
    .head(1)
)

best_suggestions.to_csv(
    "C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/club_mapping_suggestions.csv",
    index=False
)
print("Suggestions filtrées sauvegardées : club_mapping_suggestions.csv")
print(best_suggestions.head(20))


# 7️⃣ Filtrer et appliquer les correspondances de qualité (score >= 85)
mapping_filtered = best_suggestions[best_suggestions["similarity"] >= 85].copy()

# (sécurité : retirer encore une fois toute ligne sans country reconnu)
mapping_filtered = mapping_filtered[
    mapping_filtered["countries_considered"].notna() &
    (mapping_filtered["countries_considered"] != "ALL") &
    (mapping_filtered["countries_considered"].str.strip() != "")
]

mapping_filtered = (
    mapping_filtered.sort_values("similarity", ascending=False)
    .drop_duplicates("club_missing")
)

mapping_filtered.to_csv(
    "C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/club_mapping_clean.csv",
    index=False
)
print("\nFichier propre sauvegardé : club_mapping_clean.csv")
print(f"Nombre de correspondances retenues : {len(mapping_filtered)}")


# 8️⃣ Application au dataset
mapping_df = pd.read_csv(
    "C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/club_mapping_clean.csv"
)
club_mapping = dict(zip(mapping_df["club_missing"], mapping_df["suggested_match"]))

matches["HomeTeam"] = matches["HomeTeam"].replace(club_mapping)
matches["AwayTeam"] = matches["AwayTeam"].replace(club_mapping)
```

```python
# 9️⃣ Vérification finale
print("\nComplétude :")
print(matches[["HomeEloSnap", "AwayEloSnap"]].notna().mean() * 100)

missing_home = matches[matches["HomeEloSnap"].isna()]["HomeTeam"].unique()
missing_away = matches[matches["AwayEloSnap"].isna()]["AwayTeam"].unique()

print(f"\n{len(missing_home)} clubs sans Elo (Home)")
print(f"{len(missing_away)} clubs sans Elo (Away)")
```

```python
print("\nFusion des Elo manquants à partir d'EloRatings + complétude dans Matches...")

# --- Nettoyage préventif ---
matches = matches.drop(columns=[
    "HomeElo_from_elo", "AwayElo_from_elo",
    "HomeElo_from_elo_x", "HomeElo_from_elo_y",
    "AwayElo_from_elo_x", "AwayElo_from_elo_y",
    "date", "date_x", "date_y", "country_x", "country_y"
], errors="ignore")

# --- Fonction de fusion robuste ---
def merge_asof_team(matches, elo, team_col, prefix):
    merged_list = []
    for team, df_team in matches.groupby(team_col):
        df_team = df_team.sort_values("MatchDate").copy()
        df_elo = elo[elo["club"] == team].sort_values("date").copy()
        if df_elo.empty:
            df_team[[f"{prefix}Elo_from_elo"]] = pd.NA
            merged_list.append(df_team)
            continue

        df_elo = df_elo[["club", "date", "elo"]].rename(
            columns={"club": team_col, "elo": f"{prefix}Elo_from_elo"}
        )
        merged = pd.merge_asof(
            df_team,
            df_elo,
            by=team_col,
            left_on="MatchDate",
            right_on="date",
            direction="backward",
            allow_exact_matches=True
        ).drop(columns=["date"], errors="ignore")
        merged_list.append(merged)
    return pd.concat(merged_list, ignore_index=True)

# --- Fusion avec EloRatings ---
matches = merge_asof_team(matches, elo, "HomeTeam", "Home")
matches = matches.drop(columns=["date", "date_x", "date_y"], errors="ignore")
matches = merge_asof_team(matches, elo, "AwayTeam", "Away")

# Remplir les valeurs manquantes dans HomeEloSnap / AwayEloSnap 
if "HomeEloSnap" in matches.columns and "HomeElo_from_elo" in matches.columns:
    matches["HomeEloSnap"] = matches["HomeEloSnap"].fillna(matches["HomeElo_from_elo"])

if "AwayEloSnap" in matches.columns and "AwayElo_from_elo" in matches.columns:
    matches["AwayEloSnap"] = matches["AwayEloSnap"].fillna(matches["AwayElo_from_elo"])

coverage = matches[["HomeEloSnap", "AwayEloSnap"]].notna().mean() * 100
print("\nComplétude finale des Elo dans Matches :")
print(coverage)

matches = matches.drop(columns=["HomeElo_from_elo", "AwayElo_from_elo"], errors="ignore")

matches.to_csv(
    "C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/Matches_with_Elo.csv",
    index=False
)
```

```python
import random

matches_with_elo = pd.read_csv("C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/Matches_with_Elo.csv", sep=";")
elo = pd.read_csv("C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/EloRatings.csv", sep=";")

# Normaliser les dates
elo["date"] = pd.to_datetime(elo["date"], errors="coerce")
matches_with_elo["MatchDate"] = pd.to_datetime(matches_with_elo["MatchDate"], errors="coerce")

# Sélectionner 10 clubs aléatoires (avec données Elo existantes)
clubs = random.sample(matches_with_elo["HomeTeam"].dropna().unique().tolist(), 10)

print("\nVérification de cohérence Elo sur 10 clubs aléatoires :")
print("----------------------------------------------------------")

for club in clubs:
    # Sélectionner un match aléatoire pour ce club
    match_row = matches_with_elo[matches_with_elo["HomeTeam"] == club].sample(1, random_state=1)
    match_date = match_row["MatchDate"].values[0]
    match_elo = match_row["HomeEloSnap"].values[0]
    
    # Extraire l'historique Elo du club
    elo_club = elo[elo["club"] == club].copy().sort_values("date")
    
    if elo_club.empty:
        print(f"\n{club}: aucun historique trouvé dans EloRatings.csv")
        continue
    
    # Chercher le dernier Elo avant la date du match
    elo_before = elo_club[elo_club["date"] <= match_date].tail(1)
    
    print(f"\nClub : {club}")
    print(f"   Date du match : {pd.to_datetime(match_date).date()} | EloSnap utilisé : {match_elo:.1f}")
    
    if not elo_before.empty:
        real_elo = elo_before["elo"].values[0]
        delta = match_elo - real_elo
        print(f"   Elo réel (ClubElo) à cette date : {real_elo:.1f} | Différence : {delta:+.1f}")
    else:
        print("   Aucune valeur Elo trouvée avant cette date (club non suivi à ce moment).")
```

```python
import numpy as np

# Calcul automatique sur 100 clubs aléatoires
clubs = random.sample(matches_with_elo["HomeTeam"].dropna().unique().tolist(), 100)
diffs = []

for club in clubs:
    match_row = matches_with_elo[matches_with_elo["HomeTeam"] == club].sample(1)
    match_date = pd.to_datetime(match_row["MatchDate"].values[0])
    match_elo = match_row["HomeEloSnap"].values[0]
    elo_club = elo[elo["club"] == club].copy()
    elo_club["date"] = pd.to_datetime(elo_club["date"], errors="coerce")
    elo_before = elo_club[elo_club["date"] <= match_date].tail(1)
    if not elo_before.empty and not pd.isna(match_elo):
        real_elo = elo_before["elo"].values[0]
        diffs.append(match_elo - real_elo)

diffs = np.array(diffs)
print(f"\nAnalyse globale de cohérence Elo (100 clubs aléatoires)")
print(f"Moyenne des écarts : {diffs.mean():.2f}")
print(f"Écart-type : {diffs.std():.2f}")
print(f"Écart absolu moyen : {np.abs(diffs).mean():.2f}")
```

#### Création des datasets 

```python
import os

# --- Charger le dataset principal ---
df = pd.read_csv("C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/Matches_with_Elo.csv", sep=";")
df["MatchDate"] = pd.to_datetime(df["MatchDate"], errors="coerce")
df["Season"] = df["MatchDate"].dt.year

print(f"Dataset initial : {len(df)} lignes, {df.shape[1]} colonnes")

# --- Définir quelques groupes de features utiles ---
cols_target = ["FTResult"]  # variable à prédire
cols_id = ["Division", "MatchDate", "HomeTeam", "AwayTeam"]

# Colonnes post-match (à exclure pour la prédiction)
cols_post = [
    "FTHome", "FTAway", "FTResult", "HTHome", "HTAway", "HTResult",
    "HomeShots", "AwayShots", "HomeTarget", "AwayTarget",
    "HomeFouls", "AwayFouls", "HomeCorners", "AwayCorners",
    "HomeYellow", "AwayYellow", "HomeRed", "AwayRed",
    "CardsHome", "CardsAway", "CardsDiff",
    "GameDominanceIndex", "ShotsDifference", "CornersDifference"
]
cols_post = [c for c in cols_post if c in df.columns]

# --- Fonctions utilitaires ---
def filter_features(df, min_notna_ratio=0.7, exclude_postmatch=True):
    """Filtre les colonnes selon un seuil minimal de complétude."""
    na_ratio = df.isna().mean()
    keep_cols = na_ratio[na_ratio <= 1 - min_notna_ratio].index.tolist()
    if exclude_postmatch:
        keep_cols = [c for c in keep_cols if c not in cols_post]
    return keep_cols

def save_version(df, name, folder="C:/Users/Alban/Documents/IMDS/Projet_5A/Donnees/"):
    path = os.path.join(folder, name)
    df.to_csv(path, index=False)
    print(f" {name} sauvegardé ({len(df)} lignes, {df.shape[1]} colonnes)")

# --- a) matches_pred_full : toutes les années, features utilisables (≥70%) ---
cols_pred = filter_features(df, min_notna_ratio=0.7, exclude_postmatch=True)
df_pred_full = df[cols_id + cols_pred + cols_target].drop_duplicates()
save_version(df_pred_full, "matches_pred_full.csv")

# --- b) matches_recent : année seuil trouvée pour ≥80% complétude moyenne ---
min_threshold_recent = 0.8
compl_by_season = 1 - df.groupby("Season").apply(lambda x: x.isna().mean().mean())
valid_seasons_recent = compl_by_season[compl_by_season >= min_threshold_recent].index
year_min_recent = valid_seasons_recent.min() if len(valid_seasons_recent) > 0 else df["Season"].min()

df_recent = df[df["Season"] >= year_min_recent].copy()
cols_recent = filter_features(df_recent, min_notna_ratio=min_threshold_recent, exclude_postmatch=True)
df_recent_f = df_recent[cols_id + cols_recent + cols_target]
save_version(df_recent_f, "matches_recent.csv")

print(f" matches_recent : saisons ≥ {year_min_recent} (complétude moyenne ≥ {min_threshold_recent*100:.0f}%)")

# --- c) matches_clean : année seuil trouvée pour ≥90% complétude moyenne ---
min_threshold_clean = 0.9
compl_by_season_clean = 1 - df.groupby("Season").apply(lambda x: x.isna().mean().mean())
valid_seasons_clean = compl_by_season_clean[compl_by_season_clean >= min_threshold_clean].index
year_min_clean = valid_seasons_clean.min() if len(valid_seasons_clean) > 0 else df["Season"].min()

df_clean = df[df["Season"] >= year_min_clean].copy()
cols_clean = filter_features(df_clean, min_notna_ratio=min_threshold_clean, exclude_postmatch=True)
df_clean_f = df_clean[cols_id + cols_clean + cols_target]
save_version(df_clean_f, "matches_clean.csv")

print(f" matches_clean : saisons ≥ {year_min_clean} (complétude moyenne ≥ {min_threshold_clean*100:.0f}%)")

# --- d) matches_select_championnats : auto (saisons/divisions/colonnes ≥80%) ---
THRESHOLD_COL = 0.8
THRESHOLD_SEASON = 0.8
THRESHOLD_DIV = 0.8

# Moyenne par saison et division
na_by_season = 1 - df.groupby("Season").apply(lambda x: x.isna().mean().mean())
valid_seasons = na_by_season[na_by_season >= THRESHOLD_SEASON].index.tolist()
na_by_div = 1 - df.groupby("Division").apply(lambda x: x.isna().mean().mean())
valid_divs = na_by_div[na_by_div >= THRESHOLD_DIV].index.tolist()

df_select = df[df["Season"].isin(valid_seasons) & df["Division"].isin(valid_divs)].copy()
na_ratio_cols = df_select.isna().mean()
valid_cols = na_ratio_cols[na_ratio_cols <= 1 - THRESHOLD_COL].index.tolist()
valid_cols = [c for c in valid_cols if c not in cols_post]

df_select = df_select[cols_id + valid_cols + cols_target].drop_duplicates().reset_index(drop=True)
save_version(df_select, "matches_select_championnats.csv")

# --- Résumé général ---
summary = pd.DataFrame({
    "Dataset": ["matches_pred_full", "matches_recent", "matches_clean", "matches_select_championnats"],
    "Seuil complétude": ["≥70%", "≥80% (auto années)", "≥90% (auto années)", "≥80% (auto ligues/colonnes)"],
    "Année min": ["Toutes", year_min_recent, year_min_clean, sorted(valid_seasons)[0] if valid_seasons else "Toutes"],
    "Nb lignes": [len(df_pred_full), len(df_recent_f), len(df_clean_f), len(df_select)],
    "Nb colonnes": [df_pred_full.shape[1], df_recent_f.shape[1], df_clean_f.shape[1], df_select.shape[1]],
    "Complétude moyenne (%)": [
        100 * (1 - df_pred_full.isna().mean().mean()),
        100 * (1 - df_recent_f.isna().mean().mean()),
        100 * (1 - df_clean_f.isna().mean().mean()),
        100 * (1 - df_select.isna().mean().mean())
    ]
}).round(1)

print("\n Résumé des datasets générés :")
print(summary)
```
